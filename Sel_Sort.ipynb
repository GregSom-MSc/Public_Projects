This is a simple Selection Sort Algorithm. This method has a time complexity of $O(n^2)$. <br>
This makes it less efficient for bigger datasets compared to algorithms like Quick Sort or Merge Sort.

Let's import the necessary libraries.

```python
import random
import time
import matplotlib.pyplot as plt
import numpy as np
```

We add a function to measure the time it takes for the sorting function to work: 

```python
def time_code_execution(func, *args, **kwargs):
    """
    Measures a precise execution time of any given function.
    It returns a tuple containing:
        - The return value of the function.
        - The execution time in seconds.
    """
    
    start_time = time.perf_counter()  # Use perf_counter for more accurate timing
    result = func(*args, **kwargs)  # Execute the function
    end_time = time.perf_counter()

    execution_time = end_time - start_time
    return result, execution_time
```

We create a small function for sorting. the algorithm does the following:
1. Loops through each element in the array
2. Assumes that the current index holds the minimum value
3. Finds the index of the smallest element in the unsorted part of the array
4. Swaps the smallest element found with the first element of the unsorted part

```python
def selection_sort(arr):
    for i in range(len(arr)):
        
        min_idx = i
        for j in range(i+1,len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```

### Example usage

First, we create this lists. We will populate them and see the time in a graph at the end of this exercise.

```python
y = []
x = []
```

```python
# Example 1
arr = [64,25,12,22,11]
n_elements=len(arr)

result, execution_time = time_code_execution(selection_sort, arr)
y.append(execution_time)
x.append(n_elements)

print("Original array:", arr)
print(f"Result: {result}")
print(f"Execution time: {execution_time:.6f} seconds")
```

```python
# Example 2

arr_2 = [random.randint(1, 100) for _ in range(10)]
n_elements=len(range(10))
result, execution_time = time_code_execution(selection_sort, arr_2)

y.append(execution_time)
x.append(n_elements)

print("Original array:", arr_2, n_elements)
print(f"Result: {result}")
print(f"Execution time: {execution_time:.6f} seconds")
```

```python
A = [1,2,3,4,5,6,7]
for i in A:
    l_min = 500*i
    l_max = 1000*i
    
    n_elements = random.randint(l_min, l_max)
    x.append(n_elements)
    arr_i = [random.randint(1, 200) for _ in range(n_elements)]

    result, execution_time = time_code_execution(selection_sort, arr_i)
    
    y.append(execution_time)

    print("Original array:", arr_i)
    print(f"Execution time: {execution_time:.6f} seconds")
```

```python
min_len = min(len(x), len(y))  # Find the length of the shorter array
x1 = x[:min_len]  # Truncate x
y1 = y[:min_len]  # Truncate y

plt.figure(figsize=(10,5))
plt.plot(x1, y1, label="Execution Time", color="#13315C")

# Labels and title:
plt.xlabel("Input: x", fontsize=10, color="#363636")
plt.ylabel("Output: Execution Time", fontsize=10, color="#363636")
plt.title("Our Nice Execution Time line", fontsize=16, color="#363636")

# Main Borders:
plt.axhline(0,color="black", linewidth=0.5, linestyle='--')
plt.axvline(0,color="black", linewidth=0.5, linestyle='--')
plt.axhline(0.5,color="black", linewidth=0.5, linestyle='--')

# Legend:
plt.legend(fontsize=10,labelcolor="#363636")

# Grid
plt.grid(alpha=0.2)
plt.xlim(1)
plt.ylim(4.000001692678779e-06,1)
# plt.yscale('log')
# Now we show it:
plt.show
```
