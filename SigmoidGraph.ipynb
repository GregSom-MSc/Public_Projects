# Sigmoid graph
SGMS 2024

Libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Function:

```python
def sigmoid(x):
    return 1 / (1+np.exp(-x))
```

```python
# Defining the domain:
x = np.linspace(-15,15,300)
```

Let's plot it!

```python
plt.figure(figsize=(10,5))
plt.plot(x, sigmoid(x), label="Sigmoid", color="#13315C")

# Labels and title:
plt.xlabel("Input: x", fontsize=10, color="#363636")
plt.ylabel("Output: y = sigmoid(x)", fontsize=10, color="#363636")
plt.title("Our Nice Sigmoid - Activation Function", fontsize=16, color="#363636")

# Main Borders:
plt.axhline(0,color="black", linewidth=0.5, linestyle='--')
plt.axvline(0,color="black", linewidth=0.5, linestyle='--')
plt.axhline(0.5,color="black", linewidth=0.5, linestyle='--')

# Box to annotate frontier crossing point:
plt.annotate("(0,0.5)", 
             xy=(0, 0.5), xycoords='data',  # Point to annotate
             xytext=(-40, 15), textcoords='offset points',  # Offset text
             arrowprops=dict(arrowstyle="->", color="black"),  # Arrow style
             bbox=dict(boxstyle="round,pad=0.3",edgecolor="black", facecolor="#C7DCEB"),color="#363636", fontsize=8)  # Text box

# Legend:
plt.legend(fontsize=10,labelcolor="#363636")

# Grid
plt.grid(alpha=0.2)

# Now we show it:
plt.show
```
