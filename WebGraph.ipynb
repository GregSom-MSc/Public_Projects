Import the libraries:

```python
import networkx as nx
import matplotlib.pyplot as plt
```

Function:

```python
def create_radar_chart(categories, data1, data2, title="Radar Chart"):
    """Radar chart that compares two datasets.

    Args:
        categories: strings representing categories (these will be the labels around the figure).
        Data1: numerical values for dataset a.
        Data2: numerical values for dataset b.
    """

    num_categories = len(categories)

    # Angle for each axis (in radians)
    angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False).tolist()
    angles += angles[:1]  # Close the plot

    # Initialize plot
    fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(polar=True))

    # Plot each dataset
    ax.plot(angles, data1 + data1[:1], marker='o', label='Data Set 1', color='#FFC300')  # Connect the first and last points
    ax.fill(angles, data1 + data1[:1], alpha=0.25, color='#FFD60A')  # Fill area

    ax.plot(angles, data2 + data2[:1], marker='o', label='Data Set 2', color='#003566')
    ax.fill(angles, data2 + data2[:1], alpha=0.25, color='blue')

    # Set labels and title
    ax.set_xticks(angles[:-1])  # Exclude the last tick (which is a duplicate)
    ax.set_xticklabels(categories)
    ax.set_title(title, fontsize=14, color="#363636", fontweight="bold", pad=20)

    
    
    # We can further customize the appearance:
    ax.grid(True, alpha=0.5)
    ax.set_rlabel_position(0.2)  # Moving radial labels away from origin
    ax.set_yticklabels([])  # Removing radial tick labels (optional)
    ax.legend(loc=(0.9, 0.85), labelcolor="#363636")  # Placing legend outside the plot area

    plt.tight_layout()
    plt.show()
```
